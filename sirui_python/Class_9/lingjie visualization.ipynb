**Importing Matplotlib**

```python
import matplotlib.pyplot as plt
# We will use plt most often
import numpy as np
```

**A Simple Example**

```python
#plt.plot(x,y), x代表横轴用什么数据，y代表纵轴用什么数据
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

**We can use another way to create a same plot**

```python
fig1 = plt.figure() #create a figure
ax1 = fig1.add_subplot() #add a subplot(axes) to the figure we just created
ax1.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

在我们刚刚的试验中，2种方法产生了一模一样的效果，可是语法明显不一样，它们的区别是什么？
1. 第一种方法(plt.plot)是帮我们在后台自动创建了一个画布，随后在这块画布内自动选定了一整个区域作图,我们把它叫作scripting level
2. 第二种看似更复杂的方法实际上是在第一种方法上增加了更多的控制和自定义。我们可以在这种方法上给画布命名（fig1），也可以给这个画布内我们选定的区域命名（ax1)。 我们把这个方法称为artist level。

两种方法我们应该怎么去选择呢？
1. 当我们想要快速画图，且预期不会对图画细节作出大量复杂调整时，plt.plot是个好的选择
2. 当我们想要对画布，画布内的区域，及后续作图的细节作出大量的调整时，强烈推荐第二种方法。

*Notes: plt.plot()是matplotlib为了原来MATLAB的用户所创建的一个快捷的方法。
Matplotlib官方推荐使用第二种方法，体现了Python的核心--面向对象(object-oriented)。在我们后续的学习中，我们也会主要围绕第二种方法，

**Parts of figures**

**Figure** 可以解释为画布  
fig = plt.figure()  
画图的第一件事，就是创建一个画布figure，然后在这个画布上加各种元素。它为最大的一个container，大家想象成我们给观众展现我们的画作时，是展示整块画布的，而不是画布的某个地方。  
  
**Axes** 可以解释为坐标系（不是坐标轴！！！）  
ax = fig.add_subplot(1，1，1)  
在画布内，我们可以划定区域作画，默认为1，即整块画布我们画同一幅图。当我们想要在画布内划出多个区域分别作画时，比如上下各2块区域，定义方法为 ax = fig.add_subplot(2,2,1)。其中前2个2告诉python，我们想要一个2x2的区域，而且我们现在想要在第一个区域（左上角）作画。  
*axes(subplot)几乎包含了所有我们可能会对图形细节作出修改的参数*  
  
**Axis**这才是真正的坐标轴，包含了xaxis和yaxis

**Set the color, styles and markers**

```python
# Set up the line style
x = np.linspace(0, 10, 100)
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
#plt.plot(x, x + 0, linestyle='-') # solid
#plt.plot(x, x + 1, linestyle='--') # dashed
#plt.plot(x, x + 2, linestyle='-.') # dashdot
#plt.plot(x, x + 3, linestyle=':'); # dotted
```

```python
# Set up the color
plt.plot(x, x+1, color='blue') # specify color by name
plt.plot(x, x+2, color='g') # short color code (rgbcmyk)
plt.plot(x, x+3, color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
```

一般来说，matplotlib默认的8种简易颜色已经基本可以满足大部分的作画需求，8种简易色为  
'b' as blue  
'g' as green  
'r' as red  
'c' as cyan  
'm' as magenta  
'y' as yellow 
'k' as black  
'w' as white  
**更多颜色名字如下**

```python
# We can also simply change the color and linestyle at the same time
plt.plot([1, 2, 3, 4], [1, 4, 2, 3],color='green',linestyle='--')
# Matplotlib also support the below syntax to simplify the code, whch is exactly the same as the above example
plt.plot([1, 2, 3, 4], [1, 4, 2, 3],'g--')
```

```python
# When we plot the line graph, you will notice that subsequent points are linearly interpolated by default. We can set
#up the marker option to highlight the actual number in the data
plt.plot([1, 2, 3, 4], [1, 4, 2, 3],marker='o')
```

**Figure and Axes**

```python
# A common way to put multiple axes in the same figure?
fig1, axes1 = plt.subplots()  # a figure with a single Axes
fig2, axes2 = plt.subplots(2,2)  # a figure with a 2x2 grid of Axes
```

```python
fig1
```

```python
fig2 #我们通过命名figure的形式可以很轻松地调用任何之前我们定义的figure
# You may notice that the yaxis ticks are too close to the adjacent subplots, how should we adjust the distance 
# to avoid the overlapping?
```

```python
# Now let's try draw something on the 2x2 grid. Please notice the axes index, it is like Numpy two-dimensional array
x = [2,3,4,5,6]
y = [1,5,9,13,17]
fig3, axes3 = plt.subplots(2,2)
axes3[0,0].plot(x,[y-4 for y in y])
axes3[0,1].plot(x,[x**3 for x in x],'r--')
axes3[1,0].scatter(x,y,color='purple')
axes3[1,1].bar(x,y,color='k')
#此处我们的语法是ax.plot().通过指定具体的axes，我们可以在每个画布内的区域画图，而且这些区域都属于名字叫作fig3的画布
#由于plot默认是line graph，所以要画别的图形时请用相对应的method,比如bar，scatter，pie，hist
```

```python
fig3
```

**Adjust the Axis Limit**

```python
# Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes
# it’s nice to have finer control.

#调用我们第一次设立的figure (fig1),并且在fig1中的区域ax处画图
x = np.linspace(0, 10, 100)
y = 2*x + 3
axes1.plot(x, y,'b--')
fig1
```

```python
#使用set_xlim() and set_ylim()为某个选定的axes设置坐标的上下限
axes1.set_xlim(-10, 110)
axes1.set_ylim(-2, 30)
fig1
```

```python
# What if we want to adjust the x-axis limit for a particular axes in the figure? 

x = [2,3,4,5,6]
y = [1,5,9,13,17]
fig, axes = plt.subplots(2,2)
axes[0,0].plot(x,[y-4 for y in y])
axes[0,1].plot(x,[x**3 for x in x],'r--')
axes[1,0].scatter(x,y,color='purple')
axes[1,1].bar(x,y,color='k')
axes[0,0].set_xlim(-1,20) #xlim() does not work here, we will see a summary regarding the plt method and axes method later
axes[1,1].set_xlim(1,15) #xlim() does not work here, we will see a summary regarding the plt method and axes method later
```

**Setting the title, axis labels and ticks**

```python
# generate the sample data
arr = np.random.randn(1000)
```

```python
plt.plot(arr.cumsum()) #Plt menthod can simply help us get the sense of how data looks like
```

```python
plt.plot(arr)
```

```python
fig,ax = plt.subplots() #create a figure with 1 axes
ax.plot(arr.cumsum(),label='Cumulative Sum')
ax.plot(arr,label='number') #在同一个axes中再次作画
```

```python
ax.set_title('My first matplotlib plot') #设置图标的名称
ax.set_xlabel('Stages') #设置x轴的名字
ax.set_ylabel('Number') #设置y轴的名字
ax.set_xticks([0, 250, 500, 750, 1000]) #让x轴的显示数字变为list里面的数字
ax.legend(loc='best') #如果有多个graph，legend可以说明每个graph的含义 (lower left/right, upper left/right, best)
fig
```

**虽然大多数的plt函数都可以直接转换为ax的方法进行调用（例如plt.plot() → ax.plot()，plt.legend() → ax.legend()等），但是某些用于设置limits、labels和title的方法都有一定的改变。下表列出了将 MATLAB 风格的函数转换为面向对象的方法的区别.**    
    • plt.xlabel() → ax.set_xlabel()  
    • plt.ylabel() → ax.set_ylabel()  
    • plt.xlim() → ax.set_xlim()  
    • plt.ylim() → ax.set_ylim()  
    • plt.title() → ax.set_title()
