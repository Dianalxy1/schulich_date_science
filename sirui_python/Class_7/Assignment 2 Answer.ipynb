```python
import pandas as pd
df_c=pd.read_excel('/Users/sjiang1009/OneDrive - officepro449/SavvyPro/Python/data/city_edition_mapping.xlsx')
df_o=pd.read_csv('/Users/sjiang1009/OneDrive - officepro449/SavvyPro/Python/data/olympics.csv')
df_c
df_o
```

```python
df=pd.merge(left=df_o,right=df_c,on='Edition',how='inner')
df

usa_beijing = df[(df.City=='Beijing') & (df.NOC=="USA") & (df.Discipline == "Swimming")]
#Q1: 
print("USA Beijing Swimming # of Medals: ", usa_beijing.shape[0])

#Q2: 
print("USA Beijing Swimming # of Golds: ", usa_beijing[usa_beijing.Medal == "Gold"].shape[0])

#Q3: 
print(usa_beijing.Event.value_counts())

#Q4:
beijing = df[(df.City=='Beijing')]
medal_list = beijing.pivot_table(values='Discipline', index='NOC', columns='Medal', aggfunc='count')
medal_list.sort_values(by="Gold",ascending=False).head(20)[["Gold","Silver","Bronze"]]
```
